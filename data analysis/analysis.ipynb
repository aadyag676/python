```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
```

```python
avg_age = round(df['age'].mean())
print(avg_age)
```

```python
df['age']= df['age'].fillna(avg_age)
df.isnull().sum()
```

```python
df = df.dropna(subset=['embarked'])
df.isnull().sum()
```
